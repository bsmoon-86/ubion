```python
x = np.linspace(-5, 5, 101)
x
```

```python
y = (1 / np.sqrt(2 * np.pi)) * np.exp(- x**2/2)
y
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import scipy.stats as stats
```

```python
y_ = stats.norm(0, 1).pdf(x)
y_
```

```python
plt.plot(x, y_)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def norm(_a,_b,_x):
    return stats.norm(_a, _b).pdf(_x)

x = np.linspace(-20, 20, 101)
for i in range(1, 11, 1):
    _y = norm(i, i, x)
    plt.plot(x, _y)
plt.show()
```
