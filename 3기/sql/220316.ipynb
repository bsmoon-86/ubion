1. emp table 과 dept table을 left join 조건 DEPTNO가 같은 경우
2. 출력 컬럼은 ENAME, JOB, SAL, DEPTNO, LOC 
3. ENAME, LOC 그대로 출력
4. JOB은 각각 한글명으로 변경
5. SAL은 2450을 기준으로 2450 이상이면 HIGH 미만이면 LOW
6. DEPTNO은 DEPTNO의 값과 dept table의 DNAME 값을 더해서 표시 (  DEPTNO-DNAME 형식  )
6. 모든 컬럼의 이름은 한글로 변경

```python
sql_17 = """
            SELECT 
            emp.ENAME AS "사원명", 
            CASE 
                WHEN emp.JOB = "ANALYST" THEN "분석가"
                WHEN emp.JOB = "CLERK" THEN "점원" 
                WHEN emp.JOB = "MANAGER" THEN "관리자" 
                WHEN emp.JOB = "PRESIDENT" THEN "사장" 
                ELSE "판매사원"
            END AS "직급", 
            IF ( emp.SAL >= 2450, "HIGH", "LOW") AS "월급", 
            CONCAT(emp.DEPTNO, "-", dept.DNAME) AS "부서번호-부서명",
            dept.LOC AS "위치"
            FROM emp LEFT JOIN dept 
            ON emp.DEPTNO = dept.DEPTNO
            ORDER BY `사원명` ASC
        """
result_17 = _dt.executeAll(sql_17)
pd.DataFrame(result_17)
```

>   부서의 정보를 담고 있는 테이블

#### 부서의 이름이 'sales'인 소속 사원들의 리스트를 출력하라

```python
sql_20 = """
            SELECT 
            ENAME 
            FROM emp 
            LEFT JOIN dept
            ON emp.DEPTNO = dept.DEPTNO 
            WHERE dept.DNAME = "sales"
        """
result_20 = _dt.executeAll(sql_20)
pd.DataFrame(result_20)
```

```python
sql_21 = """
            SELECT 
            ENAME 
            FROM emp 
            WHERE DEPTNO = 
            (
                SELECT DEPTNO FROM dept 
                WHERE DNAME = "sales"
            )
        """
result_21 = _dt.executeAll(sql_21)
pd.DataFrame(result_21)
```

### 부서의 위치가 'NEW YORK', 'DALLAS'가 아닌 소속 사원들의 리스트를 출력하라

```python
sql_22 = """
            SELECT 
            ENAME
            FROM emp 
            WHERE 
            DEPTNO IN 
            (
                SELECT 
                DEPTNO
                FROM dept 
                WHERE 
                LOC NOT IN ('NEW YORK', 'DALLAS')
            )
        """
result_22 = _dt.executeAll(sql_22)
pd.DataFrame(result_22)
```
